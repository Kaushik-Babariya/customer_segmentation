```python
import pandas as pd
import numpy as np
import  matplotlib.pyplot as plt
```

```python
df=pd.read_csv('train.csv')
df
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Order Date']
```

```python
df['Order Date']=pd.to_datetime(df['Order Date'],dayfirst=True)
```

```python
df['Order Date']
```

```python
df.groupby(df['Order Date'].dt.year)['Sales'].sum()
```

```python
year_wise=dict(df.groupby(df['Order Date'].dt.year)['Sales'].sum())
```

```python
years=map(str,list(year_wise.keys()))
years=list(years)
years
```

```python
values=list(year_wise.values())
values
```

```python
plt.bar(years,values)
```

```python
newdf=dict(df.groupby(df['Customer ID'])['Sales'].sum())
newdf
```

```python
finaldf=pd.DataFrame({'CustID':newdf.keys(),"Sales":newdf.values()})
```

```python
highest_spend=finaldf.sort_values(by=['Sales']).head(10)
highest_spend
```

```python
plt.figure(figsize=(10,5))
plt.barh(highest_spend['CustID'],highest_spend['Sales'])
```

```python
lowest_spend=finaldf.sort_values(by=['Sales']).tail(10)
lowest_spend
```

```python
plt.figure(figsize=(10,5))
plt.barh(lowest_spend['CustID'],lowest_spend['Sales'])
```

```python
state_dict=dict(df['State'].value_counts())
#state_dict
keys=list(state_dict)
#keys
first=(keys[0],state_dict[keys[0]])
last=(keys[-1],state_dict[keys[-1]])
print('Highest Spending :',first)
print('Lowest Spending :',last)
```

```python
plt.figure(figsize=(10,10))
plt.pie(list(state_dict.values())[:20],labels=list(state_dict.keys())[:20],autopct='%1.1f%%')
```

```python
postal_analysis=dict(df['Postal Code'].value_counts())
keys=list(postal_analysis)
first=(int(keys[0]),postal_analysis[keys[0]])
last=(int(keys[-1]),postal_analysis[keys[-1]])
print('Highest Spending :',first)
print('Lowest Spending :',last)
```

```python
postal_analysis
```

```python
postal_analysis_df=pd.DataFrame({'Postal Code':postal_analysis.keys(),'Units Sold':postal_analysis.values()})
postal_analysis_df
```

```python
postal_analysis_df['Postal Code']=postal_analysis_df['Postal Code'].astype(int)
```

```python
postal_analysis_df['Postal Code']=postal_analysis_df['Postal Code'].astype(str)
postal_analysis_df
```

```python
top10_postal=postal_analysis_df.head(10)
```

```python
plt.figure(figsize=(10,5))
plt.bar(top10_postal['Postal Code'],top10_postal['Units Sold'])
plt.show()
```

```python
low10_postal=postal_analysis_df.tail(10)
```

```python
plt.figure(figsize=(10,5))
plt.bar(low10_postal['Postal Code'],low10_postal['Units Sold'])
plt.show()
```

```python
region_dict=dict(df['Region'].value_counts())
keys=list(region_dict)
first=(keys[0],region_dict[keys[0]])
last=(keys[-1],region_dict[keys[-1]])
print('Highest Spending :',first)
print('Lowest Spending :',last)
```

```python
plt.figure(figsize=(10,10))
plt.pie(list(region_dict.values()),labels=list(region_dict.keys()),autopct='%1.1f%%')
plt.show()
```

```python
city_analysis=dict(df['City'].value_counts(ascending=True))
```

```python
city_analysis
```

```python
plt.figure(figsize=(10,15))
plt.barh(list(city_analysis.keys())[-40:],list(city_analysis.values())[-40:])
for index,value in enumerate(list(city_analysis.values())[-40:]):
    plt.text(value,index,
            str(value))
plt.show()   
```
